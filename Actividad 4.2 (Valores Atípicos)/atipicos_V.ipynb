```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

__Primero, importo mi archivo 'Sin_Nulos_Valencia.csv' en el DF. Después, obtengo info del DF. Posterior, clasifico las variables en cualitativas con object y cuantitativas con int y float, concluido esto, genero la visualización de los valores en el diagrama de bigote.Sigo con la verificación de los datos correctos.__
 


```python
data= pd.read_csv('Sin_Nulos_Valencia.csv')
```

```python
data.info()
```

```python
cualitativas = data.select_dtypes(include=['object'])
cuantitativas = data.select_dtypes(include=['int64', 'float64'])
```

```python
fig= plt.figure(figsize=(15,8))
cuantitativas.plot(kind='box', vert=False)
plt.title("Valores Atipicos del df de Valencia")
plt.show()
```

```python
cuantitativas.info()
```

__Aquí, inicio con la parte de Desviación Estandar (DES), para generar los limites que tendrá "y". Después de esto salen los valores atipicos y estos los relleno con la media para después confirmar que ya no existen valores nulos. Saco por un ultimo paso los datos para crear el diagrama de DES, teniendo diferencias con el original__

```python
y=cuantitativas
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferir= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Superior)
```

```python
data3= cuantitativas[(y<=Limite_Superior) & (y>=Limite_Inferir)]
data3
```

```python
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
data_clean=data3.copy()
data_clean=data_clean.fillna(round(data3.mean(),1))
data_clean
```

```python
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
cuantitativas_DES=pd.concat([data_clean],axis=1)
cuantitativas_DES
```

```python
fig= plt.figure (figsize=(15,8))
cuantitativas_DES.plot(kind="box", vert=False)
plt.title("Valores Atípicos de Valencia con Desviación Estandar")
plt.show()
```

__Ahora paso para quitar los valores atipicos mediante IQR, generando los limites por cuartiles. Después de definidos obtenemos los valores atipicos, vemos que son distintos a los generados por DES, hago el relleno por promedio de estos valores nulos, despues compruebo que no existan para llegar al punto donde los llamo para mostrar el diagrama de IQR, siendo diferente a las dos antes mostradas del DES y el DF inicial.

```python
y=cuantitativas
percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q2
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitidos", Limite_Superior_iqr)
print("Limite inferir permitido", Limite_Inferior_iqr)
```

```python
data3_iqr= cuantitativas[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
data4_iqr=data3_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data3_iqr.mean(),1))
```

```python
Datos_limpios=pd.concat([cualitativas, data4_iqr], axis=1)
Datos_limpios
```

```python
valores_nulos=Datos_limpios.isnull().sum()
valores_nulos
```

```python
cuantitativas_IQR=pd.concat([Datos_limpios],axis=1)
cuantitativas_IQR
```

```python
fig= plt.figure (figsize=(15,8))
cuantitativas_IQR.plot(kind="box", vert=False)
plt.title("Valores Atípicos de Valencia con IQR")
plt.show()
```

__Concluyo con la elaboración de los dos csv de cada tipo de eliminación de atipicos.__
